```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
import pandas as pd
```

```python
gas = pd.read_csv('gas_turbines.csv')
gas.head()
```

```python
gas['TEYCtg'] = pd.cut(gas.TEY,bins=[100,140,180],labels=["Low","High"])
gas
```

```python
from sklearn import preprocessing 
label_encoder = preprocessing.LabelEncoder()
```

```python
gas['TEYCtg']= label_encoder.fit_transform(gas['TEYCtg']) 
gas
```

```python
Y = gas.iloc[:,11]
Y
```

```python
#using ambient variables as features i.e. predictors
X = gas.iloc[:,0:3]
X
```

```python
model = Sequential()
model.add(Dense(12, input_dim=3, kernel_initializer='random_uniform', activation='relu'))
model.add(Dense(8, kernel_initializer='random_uniform', activation='relu'))
model.add(Dense(1, kernel_initializer='random_uniform', activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
def norm_func(i):
     x = (i-i.min())	/	(i.max()	-	i.min())
     return (x)
```

```python
predictors = gas.iloc[:,0:3]
target = gas.iloc[:,11]
target
```

```python
predictors1 = norm_func(predictors)
data = pd.concat([predictors1,target],axis=1)
data
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(predictors1,target, test_size=0.3)
```

```python
def prep_model(hidden_dim):
    model = Sequential()
    for i in range(1,len(hidden_dim)-1):
        if (i==1):
            model.add(Dense(hidden_dim[i],input_dim=hidden_dim[0],activation="relu"))
        else:
            model.add(Dense(hidden_dim[i],activation="relu"))
    model.add(Dense(hidden_dim[-1],kernel_initializer="normal",activation="sigmoid"))
    model.compile(loss="binary_crossentropy",optimizer = "rmsprop",metrics = ["accuracy"])
    return model 
```

```python
first_model = prep_model([3,1])
first_model.fit(np.array(x_train),np.array(y_train),epochs=150)
pred_train = first_model.predict(np.array(x_train))
```

```python
scores = model.evaluate(X, Y)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100)) 
```

```python
pred_train = pd.Series([i[0] for i in pred_train])
```

```python
size = ["Low","High"]
pred_train_class = pd.Series(["Low"]*10527)
pred_train_class[[i>0.5 for i in pred_train]]= "High"
```

```python
train = pd.concat([x_train,y_train],axis=1)
train["TEYCtg"].value_counts()
```

```python
from sklearn.metrics import confusion_matrix
train["original_class"] = "Low"
train.loc[train["TEYCtg"]==1,"original_class"] = "High"
train.original_class.value_counts()
confusion_matrix(pred_train_class,train["original_class"])
np.mean(pred_train_class==pd.Series(train["original_class"]).reset_index(drop=True)) 
pd.crosstab(pred_train_class,pd.Series(train["original_class"]).reset_index(drop=True))
```

```python
pred_test = first_model.predict(np.array(x_test))
pred_test = pd.Series([i[0] for i in pred_test])
pred_test_class = pd.Series(["Low"]*4512)
pred_test_class[[i>0.5 for i in pred_test]] = "High"
test =pd.concat([x_test,y_test],axis=1)
test["original_class"]="Low"
test.loc[test["TEYCtg"]==1,"original_class"] = "High"
```
