```python
import pandas as pd 
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale 
```

```python
wine = pd.read_csv('wine.csv')
wine.head()
```

```python
wine_data = wine.iloc[:,1:]
wine_data.head()
```

```python
WINE = wine_data.values
WINE
```

```python
wine_normal = scale(WINE)
```

```python
wine_normal
```

```python
pca = PCA(n_components = 13)
pca_values = pca.fit_transform(wine_normal)
pca_values
```

```python
pca.components_
```

```python
var = pca.explained_variance_ratio_
var
```

```python
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
finalDf = pd.concat([pd.DataFrame(pca_values[:,0:3],columns=['pc1','pc2','pc3']), wine[['Type']]], axis = 1)
finalDf
```

 ## Hierarchical Clustering of pca data                  

```python
from sklearn.preprocessing import MinMaxScaler
trans = MinMaxScaler()
data = pd.DataFrame(trans.fit_transform(finalDf.iloc[:,0:3]))
data 
```

```python
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch # for creating dendrogram 
#p = np.array(df_norm) # converting into numpy array format 
z = linkage(data, method="complete",metric="euclidean")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(
    z,
    #leaf_rotation=0.,  # rotates the x axis labels
    #leaf_font_size=8.,  # font size for the x axis labels
)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering 
h_complete = AgglomerativeClustering(n_clusters=3, linkage='complete',affinity = "euclidean").fit(data) 

cluster_labels=pd.Series(h_complete.labels_)
cluster_labels
wine['clust']=cluster_labels # creating a  new column and assigning it to new column 
wine
```

## kmeans clustering of pca data

```python
from sklearn.cluster import KMeans
fig = plt.figure(figsize=(10, 8))
WCSS = []
for i in range(1, 11):
    clf = KMeans(n_clusters=i)
    clf.fit(data)
    WCSS.append(clf.inertia_) # inertia is another name for WCSS
plt.plot(range(1, 11), WCSS)
plt.title('The Elbow Method')
plt.ylabel('WCSS')
plt.xlabel('Number of Clusters')
plt.show()
```

```python
clf = KMeans(n_clusters=3)
y_kmeans = clf.fit_predict(data)
```

```python
y_kmeans
clf.labels_
```

```python
cd = pd.Series(y_kmeans)
winek = pd.read_csv('wine.csv')
winek['clust'] = cd
winek
```

```python
winek.groupby(['clust']).size() #kmeans clustering
```

```python
wine.groupby(['clust']).size() #hierarcical clustering
```

```python
wine.groupby(['Type']).size() #og data clusters i.e. type
```

#### It is evident that kmeans clustering is more accurate than hierarchical clustering
